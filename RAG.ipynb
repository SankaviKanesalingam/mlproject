```python
import fitz  # PyMuPDF

def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text("text") + "\n"
    return text

# Upload the PDF
from google.colab import files
uploaded = files.upload()
pdf_path = list(uploaded.keys())[0]

# Extract text
pdf_text = extract_text_from_pdf(pdf_path)
```

```python
from langchain.schema import Document
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Convert extracted text into a LangChain Document
documents = [Document(page_content=pdf_text)]

# Split into chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
chunks = text_splitter.split_documents(documents)

print(f"Total Chunks: {len(chunks)}")
print(chunks[0].page_content)  # Print the first chunk
```

Retrieve Documents Based on a Query

```python
query = "What are the benefits of renewable energy?"
retrieved_docs = vector_db.similarity_search(query, k=3)

for doc in retrieved_docs:
    print(doc.page_content)
```

```python
llm = OpenAI(model_name="gpt-4")
qa_chain = RetrievalQA(llm=llm, retriever=vector_db.as_retriever())

response = qa_chain.run(query)
print(response)
```

```python
import fitz  # PyMuPDF

def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text("text") + "\n"
    return text

# Extract text
pdf_text = extract_text_from_pdf(pdf_path)

# Print first 1000 characters to verify
print(pdf_text[:1000])
```

```python
from langchain.schema import Document
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Wrap extracted text in Document format
documents = [Document(page_content=pdf_text)]

# Split text into chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
chunks = text_splitter.split_documents(documents)

print(f"Total chunks created: {len(chunks)}")
```

```python
from langchain_openai import OpenAIEmbeddings
from langchain.vectorstores import FAISS

# Create embeddings
embeddings = OpenAIEmbeddings()

# Store chunks in FAISS
vector_db = FAISS.from_documents(chunks, embeddings)

# Save FAISS index locally in Colab
vector_db.save_local("faiss_index")

print("FAISS index saved successfully.")
```
